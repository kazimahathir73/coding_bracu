```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
from numpy.polynomial import Polynomial
```

```python
#question-01

#part a
coeff = [2.0,3.0,5.0,0.0,0.0,7.0]
degree = len(coeff)-1
print('degree-',degree)

#part b

p = Polynomial(coeff)
display(p)

#part c
def polyn(x):
  last_poly = 0
  for i in range(len(coeff)):
    sum = coeff[i] * (x**i)
    last_poly += sum
  return last_poly
x = 1.5
p_x = polyn(x)
print('The value of the function at x=1.5',p_x)

#part d
first_derivative = p.deriv()
print('First derivative:',first_derivative)

#part e
x = 1.5
first_derivative_x = first_derivative(x)
print('First derivative:',first_derivative_x)
```

```python
#question 2
data_x = [1,2,3]
M = np.zeros((len(data_x),len(data_x)))

for i in range(len(M)-1,-1,-1):
  deg = len(data_x) - 1
  for k in range(len(M[i])):
    M[i][k] = data_x[i]*deg
    deg -= 1

print(M)
```
