# Problem-01

```python
class Student:
    stu_id=0
    
    def __init__(self,name,dep,age,cg):
        self.name=name
        self.dep=dep
        self.age=age
        self.cg=cg
        Student.stu_id+=1
        
    def get_details(self):
        print('ID:',Student.stu_id)
        print('Name:',self.name)
        print('Department:',self.dep)
        print('Age:',self.age)
        print('CGPA:',self.cg)
        
    @classmethod
    def from_String(cls,info):
        cls.list=info.split('-')
        return Student(cls.list[0],cls.list[1],cls.list[2],cls.list[3])

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

print('Instance veriable had to call by an object and class veriable had to call by the class')
print("Instance method need a class instance and can access the instance through self but Class methods don't need a class instance.They can't access the instance ( self ) but they have access to the class itself via cls")
```

# Problem-02

```python
class Assassin:
    num=0
    
    def __init__(self,name,damage):
        self.name=name
        self.damage=damage
        Assassin.num+=1
        
    def printDetails(self):
        print('Name:',self.name)
        print('Success rate:',self.damage)
        print('Total number of Assassin:',Assassin.num)
        
    @classmethod
    def failureRate(cls,name,percent):
        cls.name=name
        cls.percent=100-percent
        return Assassin(cls.name,cls.percent)
    
    @classmethod
    def failurePercentage(cls,name,percent):
        cls.name=name
        cls.percent=100-percent
        return Assassin(cls.name,cls.percent)

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
class Passenger:
    count=0
    
    def __init__(self,name):
        self.name=name
        self.price=450
        Passenger.count+=1
        
    def set_bag_weight(self,weight):
        if weight > 20  and weight < 50:
            self.price+=50
        else:
            self.price+=100
            
    def printDetail(self):
        print('Name:',self.name)
        print('Bus Fare:',self.price)
        
print('Total Passenger:', Passenger.count)
p1 = Passenger('Jack')
p1.set_bag_weight(90)
p2 = Passenger('Carol')
p2.set_bag_weight(10)
p3 = Passenger('Mike')
p3.set_bag_weight(25)
print("=========================") 
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================") 
print('Total Passenger:', Passenger.count)
```

# Problem-05

```python
class Employee:
    def __init__(self,name,wp):
        self.name=name
        self.workingPeriod=wp
    
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        cls.year=2021-year
        return Employee(name,cls.year)
    
    @staticmethod
    def experienceCheck(year,gender):
        if gender=='male' and year < 3:
            return 'He is not experienced'
        elif gender=='male' and year >= 3:
            return 'He not experienced'
        elif gender=='female' and year < 3:
            return 'She is not experienced'
        elif gender=='female' and year >= 3:
            return 'She is experienced'

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Laptop:
    laptopCount=0
    
    def __init__(self,brand,count):
        self.name=brand
        self.count=count
        Laptop.laptopCount+=self.count
    
    @classmethod
    def advantage(cls):
        print('Laptops are portable')
        
    @classmethod
    def resetCount(cls):
        Laptop.laptopCount=0

lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

# Problem-07

```python
class Cat:
    Number_of_cats=0
    
    def __init__(self,color='White',action='Sitting'):
        self.color=color
        self.action=action
        Cat.Number_of_cats+=1
        
    def printCat(self):
        print(self.color,'cat is',self.action)
        
    def changeColor(self,color1):
        self.color=color1
        
    @classmethod
    def no_parameter(cls):
        return Cat('White','Sitting')
        
    @classmethod
    def first_parameter(cls,color):
        cls.color=color
        return Cat(color)
    
    @classmethod
    def second_parameter(cls,action):
        return Cat(cls.color,action)

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue") 
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

# Problem-08

```python
class Cylinder:
    radius=5
    height=18
    
    def __init__(self,r,h):
        self.radius=r
        self.height=h
        print('Default radius=',Cylinder.radius,'and height=',Cylinder.height)
        print('Updated: radius=',self.radius,'and height=',self.height)
        Cylinder.radius=self.radius
        Cylinder.height=self.height
        
    @staticmethod
    def area(r,h):
        area=(2*3.1415926535897932384626433832795*Cylinder.radius**2)+(2*3.141592653589793238462643383*Cylinder.radius*Cylinder.height)
        print('Area:',area)
        
    @staticmethod
    def volume(r,h):
        volume=3.141592653589793238462643383*Cylinder.height*Cylinder.radius**2
        print('Volume:',volume)
        
    @classmethod
    def swap(cls,h,r):
        return Cylinder(r,h)
    
    @classmethod
    def changeFormat(cls,info):
        list1=info.split('-')
        return Cylinder(int(list1[0]),int(list1[1]))

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class Student:
    t_stu=0
    brac_stu=0
    other_stu=0
    
    def __init__(self,name,department,institution='BRAC University'):
        self.name=name
        self.dep=department 
        self.int=institution
        Student.t_stu+=1
        if self.int =='BRAC University':
            Student.brac_stu+=1
        else:
            Student.other_stu+=1
        
    def individualDetail(self):
        print('Name:',self.name)
        print('Department:',self.dep)
        print('Institution:',self.int)
        
    @classmethod
    def printDetails(cls):
        print('Total Student(s):',Student.t_stu)
        print('BRAC University Student(s):',Student.brac_stu)
        print('Other Institution Student(s):',Student.other_stu)
        
    @classmethod
    def createStudent(cls,name,dep,ints='BRAC University'):
        return Student(name,dep,ints)

Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```

```python
class SultansDine:
    t_num=0
    t_price=0
    dict1={}
    
    def __init__(self,name):
        self.place=name
        SultansDine.t_num+=1
        SultansDine.dict1[name]=0
        
    def sellQuantity(self,num):
        sell=0
        if num < 10:
            sell = num * 300
        elif num < 20:
            sell = num * 350
        else:
            sell = num * 400
            
        self.quantity=sell
        SultansDine.t_price+=sell
        SultansDine.dict1[self.place]=sell
        
    def branchInformation(self):
        print('Branch Name:',self.place)
        print('Branch Sell:',self.quantity,'Taka')

        
    @classmethod
    def details(cls):
        print('Total Number of branch(s):',SultansDine.t_num)
        print('Total Sell:',SultansDine.t_price)
        if len(SultansDine.dict1)!=0:
            for i in SultansDine.dict1:
                print(f'Branch Name: {i}, Branch sell: {SultansDine.dict1[i]}')
                perc=(SultansDine.dict1[i]/SultansDine.t_price)*100
                print(f"Branch consists of total sell's: {perc}%")
        
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
