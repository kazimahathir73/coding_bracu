```python
class PlayerEarning:
    def __init__(self,name):
        self.name=name
        
    def calculateTotal(self,money,match=None):
        self.money=money
        if match != None:
            if match > 30:
                self.bonus=int(((5/100)*self.money)+10000)
            else:
                self.bonus=int(((5/100)*self.money))
        else:
            self.bonus = 0
            
    def printDetails(self):
        print('Player Name:',self.name)
        print('Player Season Earning without bonus:',self.money)
        print('Bonus:',self.bonus)
        print('Player Season Earning After Bonus:',self.money+self.bonus)


print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()
print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()
print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
class Processor:
    def __init__(self, model, thread, core):
        self.model = model
        self.core = core
        self.thread = thread
    def getInfo(self):
        return 'Model:'+self.model+ '\nCores:'+str(self.core)+ '\nThreads:'+ str(self.thread)
    
class Intel(Processor):
    def __init__(self,model,core,thread,price):
        super().__init__(model,thread,core)
        self.price=price
    
    def getInfo(self):
        print(super().getInfo()+'\n'+'Price: '+str(self.price))
        
class AMD(Processor):
    def __init__(self,model,core,thread,price):
        super().__init__(model,thread,core)
        self.price=price
    
    def getInfo(self):
        print(super().getInfo()+'\n'+'Price: '+str(self.price))
      
p1 = Intel("Intel i5 10th Gen",6,12,17000)
p2 = AMD("Ryzen 5 3500X",6,6,13800)
p3 = AMD("Ryzen 5 3600",6,12,16900)
print('======================')
p1.getInfo()
print('======================')
p2.getInfo()
print('======================')
p3.getInfo()
```

```python
class Student:
    def __init__(self,name,ID):
        self.name = name
        self.ID = ID
    def Details(self):
        return "Name: "+self.name+"\n"+"ID: "+self.ID+"\n"
    
class CSEStudent(Student):
    def __init__(self,name,id1,semester):
        super().__init__(name,id1)
        self.semester=semester
    def Details(self):
        return super().Details()+'Current semester: '+self.semester
    
    def addCourseWithMarks(self,*marks):
        self.list_course=[]
        self.list_mark=[]
        for i in range(0,len(marks)):
            if i%2 ==0:
                self.list_course.append(marks[i])
            else:
                if marks[i] >= 0 and marks[i] <= 50:
                    self.list_mark.append(0.00)
                elif marks[i] >= 51 and marks[i] <= 54:
                    self.list_mark.append(1.00)
                elif marks[i] >= 55 and marks[i] <= 56:
                    self.list_mark.append(1.33)
                elif marks[i] >= 57 and marks[i] <= 64:
                    self.list_mark.append(2.00)
                elif marks[i] >= 65 and marks[i] <= 69:
                    self.list_mark.append(2.3)
                elif marks[i] >= 70 and marks[i] <= 79:
                    self.list_mark.append(3.00)
                elif marks[i] >= 80 and marks[i] <= 84:
                    self.list_mark.append(3.3)
                else:
                    self.list_mark.append(4.00)
                       
    def showGPA(self):
        print(self.name,'has taken',len(self.list_course),'courses.')
        sum1=0
        for i in range(0,len(self.list_course)):
            sum1+=self.list_mark[i]*3
            print(self.list_course[i],':',self.list_mark[i])
            
        print('GPA of',self.name,'is',sum1/(len(self.list_course)*3))

Bob = CSEStudent("Bob","20301018",'Fall 2020')
Carol = CSEStudent("Carol","16301814",'Fall 2020')
Anny = CSEStudent("Anny","18201234",'Fall 2020')
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111",83.5,"CSE230",73.0,"CSE260",92.5)
Carol.addCourseWithMarks("CSE470",62.5,"CSE422",69.0,"CSE460",76.5,"CSE461",87.0)
Anny.addCourseWithMarks("CSE340",45.5,"CSE321",95.0,"CSE370",91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Transport:
    total_traveller = 0
    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare
    def __str__(self):
        s = 'Name: '+self.name+", Base fare: "+str(self.baseFare)
        return s
    
class Bus(Transport):
    def __init__(self,name,price):
        super().__init__(name,price)
        print(super().__str__())
        self.count1=0
        
    def addPassengerWithBags(self,*other):
        self.list1=[]
        self.list2=[]
        
        for i in range(0,len(other)):
            if i%2==0:
                self.list1.append(other[i])
            else:
                if other[i] <= 2:
                    self.list2.append(self.baseFare)
                elif other[i] >= 3 and other[i] <= 5:
                    self.list2.append(self.baseFare+60)
                else:
                    self.list2.append(self.baseFare+105)
                
        Transport.total_traveller+=len(self.list1)
        self.count1+=len(self.list1)
        
    def __str__(self):
        sum1=''
        for i in range(0,len(self.list1)):
            sum1+='Name: '+self.list1[i]+', Fare: '+str(self.list2[i])+'\n'
            
        return super().__str__()+'\nTotal Passenger(s): '+str(self.count1)+'\nPassenger details:\n' +sum1
    
class Train(Transport):
    def __init__(self,name,price):
        super().__init__(name,price)
        print(super().__str__())
        self.count1=0
        
    def addPassengerWithBags(self,*other):
        self.list1=[]
        self.list2=[]
        
        for i in range(0,len(other)):
            if i%2==0:
                self.list1.append(other[i])
            else:
                if other[i] <= 2:
                    self.list2.append(self.baseFare)
                elif other[i] >= 3 and other[i] <= 5:
                    self.list2.append(self.baseFare+60)
                else:
                    self.list2.append(self.baseFare+105)
                       
        Transport.total_traveller+=len(self.list1)
        self.count1+=len(self.list1)
        
    def __str__(self):
        sum1=''
        for i in range(0,len(self.list1)):
            sum1+='Name: '+self.list1[i]+', Fare: '+str(self.list2[i])+'\n'
            
        return super().__str__()+'\nTotal Passenger(s): '+str(self.count1)+'\nPassenger details:\n'+sum1

t1 = Bus('Volvo', 950)
print("=================================")
t1.addPassengerWithBags('David', 6, 'Mike', 1, 'Carol', 3)
print("=================================")
print(t1)
print("=================================")
t2 = Train('Silk City', 850)
print("=================================")
t2.addPassengerWithBags('Bob', 2, 'Simon', 4)
print("=================================")
print(t2)
print("=================================")
print('Total Passengers in Transport: ',Transport.total_traveller )
```
