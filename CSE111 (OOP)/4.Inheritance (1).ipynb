### Printing the Object 

Way 01(By default - address)

````python
class Student():
   def __init__(self,name,id):

      self.name = name
      self.id = id 


#driver code 
s1 = Student("John",1)
print(s1)   #printing the object - way 01 . __str__(self) is called

print(Student("Mary",2))   #printing the object - way 02  __str__(self) is called
````

Way 02(Configuring str function)

````python
class Student():
   def __init__(self,name,id):

      self.name = name
      self.id = id 

   def __str__(self): #This is called when we print the object. We manually over ride it and configure it to our output
     return   f"Name of student is {self.name} and id is {self.id}"



#driver code 
s1 = Student("John",1) 
print(s1)   #printing the object - way 01 

print(Student("Mary",2))   #printing the object - way 02
````

### Parent - Child relationship 

asasdad.drawio (2).png

```
Color Code: 
Text:  
1.   Black = Public = All can access
2.   Blue = Protected = Only own class and child class can access 
3.   Red = Private  = Only own class

Yellow Square = Not accessable
All other colored square = different class. (Blue = parent || Red = Child ||  Green = Grand child)

```




````python
class Parent():
 
       def __init__(self):
            
            self.road = 10       #Public - Can be access from anywhere
            self._land = 120     #Protected - Can only be accessed from this and subclasses
            self.__box = "Gold"  #Private - can only be accessed from this class 

      
       def keyforbox(self):
          print(self.__box)



class Child(Parent):

      def __init__(self):
         super().__init__()   #Calling the constructor of super class(Parent) to set the 3 instance variable of Parent class
        
         self.rooftop = "White" #Public
         self._building = 2      #Protected
         self.__locker  = "Coin" #Private
    
      def details_of_child(self):
         print(f"Child has access to {self.road} and {self._land} from parent")
         #print(f"Child does not have access to {self.__box}")                     #uncomment for error. As box is a private variable, can not be accessed outside Parent class even though it is inherited 
         print(f"Child has his own belongings such as {self.rooftop},{self._building} and {self.__locker}")

class GrandChild(Child):

    def __init__(self):
      super().__init__()
      self.road = "straight"



#----driver code-----

p1 = Parent()
print("Person ->",p1.road,p1._land)
#print(p1.__box)             #uncomment for error. As box is a private variable, can not be accessed outside Parent class
#print(p1.keyforbox())


c1 = Child()
print("Child ->",c1.rooftop,c1._building)
c1.details_of_child()
print("Child inherit from person ->",c1.road,c1._land)   #raod and _land are instance variable of object of  Parent class
#print(c1.__locker)         #uncomment for error. As locker is a private variable, can not be accessed outside Child class
#print(c1.__box)            #uncomment for error. As box is a private variable, can not be accessed outside Parent class




f1 = GrandChild()

print(f1.rooftop)
````
