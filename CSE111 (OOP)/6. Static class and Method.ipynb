### Instance Variable: 
1. Dependent on instance/object. Called from driver code through object 
2. If  we change an instance variable of an object , instance variable of the other objects stays unaffected. 
3. Declared only inside init. With self
4. When called or accessed inside class we have to use self infront of it. 


---



### Class/Static Variable: 
1. Dependent on class. Called from driver code using Class name
2. Changes for all objects in a class if static variable in one object is changed. 
3. Declared outside and above init.
4. While declaring we can not have anything infront of variable
5. While calling or accessing the static variable we  have to use class name infront of it


---






---






### Instance Method: 
1. Dependent on instance. Called from driver code using object's name 
2. Method must have self as a first parameter. 
3. Can access both static/class variable and instance variable(imp)
---






### Class Method: 
1. Dependent on class. Called from driver code using class name
2. While calling we have to put class name on front. For eg ClassName.methodName()
3. Method paramter must have cls, we must write @classmethod above the method
4. Cant access any instance variable from class method.(imp)


---



### Static Method:
1. No Method parameter by itself. 
2. Same as class method 


---


---




```
Class Name = Starts with a capital letter 

Instance Method = Starts with a small letter

Instance Variable Name = Starts with a small letter 
```
