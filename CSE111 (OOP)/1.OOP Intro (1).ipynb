### **Parameterless Constructor**

````python
class Building():

      #parameterless constructor
      def __init__(self): 
        self.color = "white"   #These are INSTANCE VARIABLE which must be declared inside constructor. 
        self.floor = 10        #Instance variable are INSTANT/OBJECT dependent so changing one variable of an object does not affect the value of the same variable in another object
                               #For example changing floor value in one object does not make any impact on other floor value of other objects.



b1 = Building()   #The moment we we write Building() an obejct is created in the memory. During object creation the constructor(__init__) is called. 
print("Original color of b1",b1.color)   #color variable of b1 is called
print("Original floor no. of b1",b1.floor)
b1.color = "red"  #b1's color is set to "red". NB: Does not change color for other objects
print("Updated color of b1",b1.color)   

print("----")

b2 = Building()  
print("Original color of b2",b2.color)
print("Original floor no. of b2",b2.floor)

b2.floor  = 12   #b2's floor no. is updated. But this does not effect  other objects b1.

print("Checking floor no. of b1",b1.floor)  #Original floor no. b1
print("Checking floor no. of b2",b2.floor)  #Showing updated floor no. b2


print("----")
print("b1 location",b1)     #Printing the memory location(address) gives the address of the 
print("b2 location",b2)
````

````python
class Building():

      #parameterized constructor
      def __init__(self,color,floor=None):
        self.color = color         #instance variable available for objects must be declared in constructor 
        self.floor = floor





b1 = Building("red",15)   #color = red , floor =15
#print(b1.color)
print(b1.floor)
b1.color = "violet"
print(b1.color)


print("----")

b2 = Building("blue",20)   #color = blue , floor = 10
print(b2.color)
print(b2.floor)

b2.floor = 10

print(b1.floor)
print(b2.floor)

print("----")
print(b1)
print(b2)
````

````python
class Building():
  
      def __init__(self,a,b): #parameterized constructor
        self.color = a           #instance variable
        self.floor = b

      def colorChange(self,newcolor): #instance method/function
        self.color = newcolor

      def floorChange(self,newfloor):
        self.floor = newfloor

      def increasefloor(self):
        self.floor+=1

      def detail(self):
        print("The color of building is " +self.color + 'and floor is ' + str(self.floor))


#=====driver code=====
b1 = Building("red",15)   #color = red , floor =15
print(b1.color)
print(b1.floor)
print(b1.colorChange("Violet")) # We are calling a instance method as we have b1._ _ _(). We are by default passing "self"(the object itself) as a parameter
print(b1.color)

b1.detail()

print("----")

b2 = Building("blue",20)   #color = blue , floor = 10
print(b2.color)
print(b2.floor)



print("----")

b2.floorChange(10)


print(b1.floor)
print(b2.floor)

b2.increasefloor()
print(b2.floor)
print(b2.detail())


print("----")
print(b1)
````

Default parameter only used when multiple constructor/function/methods will use different numbers of paramters/arguments

List --> Unknown number of parameter used instead of default parameters

Local varible --> Exists only in methods declared


Instance variable ---> Exists only in the object it is declared. Declared only if parameter pased when object creation or when variable used in MULTIPLE methods

````python
class Customer():

  def __init__(self,n):

    self.name = n
    

  def greet(self,g=None):
    #g is a local variable
    if g == None: 
      print("Hello!")
    else:
      print(f"Hello {g}!")

    
  def purchase(self,*args):
    #count is a local variable 
    count = 0
    #args =["chips", "chocolate", "orange juice"]
    count = len(args)
    print(f"{self.name} you purchased {count} item(s)")

    for i in args:
      print(i)




customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
````
