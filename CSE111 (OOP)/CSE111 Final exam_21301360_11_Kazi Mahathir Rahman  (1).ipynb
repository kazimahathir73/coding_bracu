```python
class Buyer:
    serial=0
    def __init__(self,name,id1=None):
        self.name=name
        self.id=id1
        Buyer.serial+=1
        
    def __str__(self):
        if self.id != None:
            str1='Registered Customer will get 100 tk discount'
        else:
            str1=''
        return 'Customer SL no.: '+str(Buyer.serial)+'\nBuyer Name: '+self.name+'\nRegistered Customer ID:'+str(self.id)+'\n'+str1
    
    def create_bill(self,other1,other2,other3=None):
        if other3 != None:
            bill=other1.bill+other2.bill+other3.bill
        else:
            bill=other1.bill+other2.bill
        print('Total bill:')
        print(str(bill)+' tk')
        
class Groceries:
    items= {'Packaged': {'Rice': 110, 'Atta': 40, 'Oats': 350}, 'Fruits': {'Mango': 200, 'Orange': 180}}
    def __init__(self,name,cata,kg=1):
        self.name=name
        self.cata=cata
        self.kg=kg
        self.bill=0
        for i,k in Groceries.items.items():
            for p,m in k.items():
                if self.name == p:
                    mul=int(m)*int(self.kg)
                    self.bill+=mul
                    
    def __str__(self):
        return 'Groceries Order Details:\n'+str(self.kg)+' kg '+self.name
    
    @classmethod
    def add_items(cls,cata,list1):
        for i in list1:
            sp=i.split()
            Groceries.items[cata][sp[0]]=sp[1]

buyer1 = Buyer('Ross', 'CL23')
print('1.===============================')
print(buyer1)
item1 = Groceries('Mango', 'Fruits')
print('2.===============================')
print(item1)
print('3.===============================')
print(Groceries.items)
print('4.===============================')
Groceries.add_items('Packaged',['Flour 90', 'Noodles 30', 'Soybean-Oil 150'])
print(f'Updated Item List:\n{Groceries.items}')
print('5.===============================')
item2 = Groceries('Flour', 'Packaged', 3)
print(item2)
print('6.===============================')
buyer1.create_bill(item1, item2)
print('7.===============================')
item3 = Groceries('Orange', 'Fruits', 3)
print(item3)
print('8.===============================')
item4 = Groceries('Rice', 'Packaged', 2)
print(item4)
print('9.===============================')
buyer2 = Buyer('Chandler')
print(buyer2)
print('10.==============================')
buyer2.create_bill(item1, item3, item4)
```
