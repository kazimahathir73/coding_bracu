```python
class Node:
  def __init__(self,elm,next):
    self.element = elm
    self.next = next
```

```python
class linked_list:
  def __init__(self,arr):
    self.head = Node(arr[0],None)
    self.current_node = self.head
    self.count = 0

  def create_linked_list(self,arr):
    if len(arr) <= 0:
      return 
    else:
      current_node = Node(arr[0],None)
      self.current_node.next = current_node
      self.current_node = self.current_node.next
      self.create_linked_list(arr[1:])

  def return_head(self):
    return self.head

  def alternative_linkedlist(self,head):
    if head != None:
      if self.count%2 == 0:
        print(head.element)
        head = head.next
        if head != None:
          self.alternative_linkedlist(head.next)
      self.count+=1
    else:
      return

array = [1,2,3,4,5]
obj = linked_list(array)
obj.create_linked_list(array[1:])
head = obj.return_head()
obj.alternative_linkedlist(head)
```
