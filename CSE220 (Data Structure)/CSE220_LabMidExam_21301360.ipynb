```python
class stack_score:
  def __init__(self,arr):
    self.arr = arr
    self.idx = len(self.arr)

  def push(self,element):
    if self.idx == len(self.arr):
      self.arr[self.idx-1] = element
    else:
      self.arr[self.idx] = element
    self.idx += 1

  def pop(self):
    self.idx -= 1
    var=self.arr[self.idx-1]
    return var

  def decider(arr1,arr2):
    player1 = stack_score(arr1)
    player2 = stack_score(arr2)

    total_player1 = 0
    total_player2 = 0

    idx = 0
    player1_push = 0
    player2_push = 0

    for i in range(len(arr1)-1,-1,-1):
      if idx < len(arr1):
        if arr1[i] % 2 != 0:
          add = player1.pop()
          total_player1 += add
        else:
          player1.pop()
          player1.push(arr1[i]+1)
          player1_push += 1

        if arr2[i] % 2 != 0:
          add = player2.pop()
          total_player2 += add
        else:
          player2.pop()
          player2.push(arr2[i]+1)
          player2_push += 1

      else:
        break

      idx+=1


    if total_player1 > total_player2:
      print('I am the winner')
    elif total_player1 == total_player2:
      print('Socre is equal')
      if player1_push < player2_push:
        print('Based on less number of push operation you friend win')
      else:     
        print('Based on less number of push operation you win')
    else:
      print('My friend is winner')
      

id1 = [1,8,4,0,1,5,4,6]
id2 = [1,9,2,1,0,3,6,5]
stack_score.decider(id1,id2)
```
