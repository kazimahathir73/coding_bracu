```python
#task-01(a)
def factorial(n):
  if n == 0:
    return 1
  else:
    return n*factorial(n-1)

print(factorial(3))
```

```python
#task-01(b)
def fibonacci(n):
  if n <= 2 and n > 0:
    return 1
  elif n == 0:
    return 0
  else:
    return (fibonacci(n-1) + fibonacci(n-2))
  
print(fibonacci(6))
```

```python
#task-01(d)
def power(base,n):
  if n == 0:
    return 1
  else:
    return base * power(base,n-1)

print(power(3,3))
```

```python
#task-02(b)

class Node:
  def __init__(self,elm,next):
    self.element = elm
    self.next = next
  
class node_create:
  def __init__(self,ary):
    self.head = Node(ary[0],None)
    current_head = self.head
    for i in range(1,len(ary)):
      n = Node(ary[i],None)
      current_head.next = n
      current_head = current_head.next
  def return_head(self):
    return self.head

#task - 2(a)
def sum_linkedlist(node):
  if node == None:
    return 0 
  else:
    return node.element + sum_linkedlist(node.next)

#task - 2(b)
def print_reverse(node):
  if node == None:
    return
  else:
    print_reverse(node.next)
    print(node.element,end=',')

obj1 = node_create([1,2,3,4,5,6,7,8])
head = obj1.return_head()
print(sum_linkedlist(head))
print_reverse(head)
```

```python
#task -04
def pyramid(num,k):
  if(k>num):
    return
  else:
    for i in range(k):
      print(i+1,end=" ")
    print()
    pyramid(num,k+1)

pyramid(5,1)
```

```python
#Task_4(a)
def pattern(i,j,n):
    if i <= n:
      if j<=i:
        print(j, end="")
        pattern(i,j+1,n)
        return
      print()
      pattern(i+1,1,n)
      return
n=5
pattern(1,1,n)
```

```python
#Task_4(b)
def pattern(i,j,n, flag = False):
    if i <= n:
      if flag is False:
        print(" "*(n-i),end="")
      if j<=i:
        print(j,end="")
        return pattern(i,j+1,n,True)
      print()
      return pattern(i+1,1,n, False)
    
n=5
pattern(1,1,n)
```

```python
def pyramid(num,i):
  if(i>num):
    return
  else:
    for k in range(num-i):
      print(" ",end=" ")
    for k in range(i):
      print(k+1,end=" ")
    print()
    pyramid(num,i+1)

pyramid(5,1)
```
