```python
#task -01

class selectionsort:
  def __init__(self,arr):
    self.array = arr
    self.idx = 0

  def sort(self,arr,i):
    if len(arr) == 0:
      return
    else:
      if i >= len(arr):
        return 
      else:
        min = arr[0]
        if arr[i] < arr[0]:
          min = arr[i]
        #print(min)
        self.sort(arr,i+1)
        self.array[self.idx] = min
      self.idx+=1
      self.sort(arr[1:],i+1)

  def print_array(self):
    print(self.array)

arr = [5,7,2,4,-1,0,9,4,7]
obj = selectionsort(arr)
obj.sort(arr,1)
obj.print_array()
```
