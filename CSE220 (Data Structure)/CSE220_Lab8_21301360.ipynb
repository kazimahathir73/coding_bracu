```python
# task 1

def selection_sort(a,i,j):
  if i == l and j==l:
    return -1

  if i < l-1:
    min=i

    while j < l:

      if a[j] < a[min]:
        min = j
      j=j+1
      
    if min != i:
      temp=a[i]
      a[i]=a[min]
      a[min]=temp
    selection_sort(a,i+1,i+2)
  

A=[13,25,0,-4,7,-1,18,9,-6,21]
i=0
j=i+1
l=len(A)
selection_sort(A,i,j)
print(A)
```

```python
# task 2

def insertion_sort(a,i):
  
  if i == l:
    return -1

  if i < l:
    j=i-1
    key=a[i]

    while j>=0 and key<a[j]:
      a[j+1]=a[j]
      j=j-1
      
    a[j+1]=key
    insertion_sort(a,i+1)



A=[22,5,14,2,7,1]
i=1
l=len(A)
insertion_sort(A,i)
print(A)
```

```python
# task 6
def binary_search(val,a,start,end):
    mid=(start+end)//2
    try:
      if a[mid]==val:
        print('Found')

      elif mid==len(a)-1 or mid==0:
        print('Not found')

      if val<a[mid]:
        binary_search(val,a,start,mid)
      else:
        binary_search(val,a,mid,end)
    except:
      print('Not found')

binary_search(6,[1,3,5,7,9,10,16,25,36,47,58,69],0,10)
```

```python
# task 7

A = []
def Fib_add(n):
  if(A[n] > 0):
    return A[n]
  if(n == 0 or n==1):
    return 1
  else:
    return A[n].append(Fib_add(n-1)+Fib_add(n-2))

def memoization(n):
  A.insert(0,1)
  A.insert(1,1)
  for i in range(2,n+1):
    A.insert(i,A[i-1]+A[i-2])
  return Fib_add(n)

print(memoization(8))
```
